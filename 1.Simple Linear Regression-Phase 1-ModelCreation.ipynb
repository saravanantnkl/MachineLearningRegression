```python
#Pandas is a libraries. it containes multiple functions and its useful to access database tables and files
import pandas as pd
```

```python
#this dataset is used for store the data from pandas object. pandas object have data from file or database table
dataset=pd.read_csv("Salary_Data.csv")
```

```python
dataset
```

```python
independent=dataset[["YearsExperience"]]#this variable is used for separate the data from dataset based on columns
independent
```

```python
dependent=dataset[["Salary"]]
dependent
```

```python
#from keyword used for to load the particular methods from that library. sklearn library is used for machine learning
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression #load particular LinearRegression class from sklearn library
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
weight=regressor.coef_ #coef_ this is assigned variable declared in LinearRegression class
weight
```

```python
bais=regressor.intercept_ #intercept_ this is assigned variable declared in LinearRegression class
bais
```

```python
y_pred=regressor.predict(x_test) #we always pass to input test data to predict the result
```

```python
#r2_score class used for to validate the r score. if score is nearly 1 then its good predict method. if score is nearly 0 then its bad predict method
from sklearn.metrics import r2_score 
r_score=r2_score(y_test,y_pred)
r_score
```

```python
#saving the model to use one of library. this pickle library used to save the model
import pickle
filename="finalized_model_linear.sav" #file extension of pickle file as .sav
```

```python
pickle.dump(regressor,open(filename,'wb'))
```

```python
load_model=pickle.load(open("finalized_model_linear.sav",'rb'))
result=load_model.predict([[15]])
```
